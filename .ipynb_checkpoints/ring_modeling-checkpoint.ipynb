```python
# Try to model a SiN based ring resonator
import numpy as np
import math
import matplotlib.pyplot as plt

# Basic parameters
n_SiN = 2
n_eff = 1.6

r = 100 # um
circ = 2*np.pi*r
wavel = np.arange(1450, 1640, 0.001) # nm
delta = np.divide(circ * n_eff * np.ones(len(wavel)), wavel/1000)

alpha_dB = 0.1 # dB/cm
alpha = alpha_dB / (10*np.log10(math.exp(1))) # /cm
a = math.exp(-alpha*circ*1e-4)
print(a)

t = 0.9985

# calculate the tranmission
P_trans = []
for phi in delta:
    trans = (a*a + t*t - 2*a*t*math.cos(phi)) / (1 + a*a*t*t - 2*a*t*math.cos(phi))
    P_trans.append(trans)
```

```python
# plot the transmission spectrum
fig, ax = plt.subplots(1, 1, figsize=(7,5))
ax.plot(wavel, 10*np.log10(P_trans))
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Transmission (dB)')
```
