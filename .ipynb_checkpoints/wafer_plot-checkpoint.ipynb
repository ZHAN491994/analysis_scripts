```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import matplotlib as mpl
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

def func(x, a):
    return np.exp(-x * a)


x = np.linspace(0, 5)

fig, ax = plt.subplots()

# define color map
# cmap = cm.get_cmap("Spectral")
cmap = mpl.colormaps['viridis']

# need to normalize because color maps are defined in [0, 1]
norm = colors.Normalize(0, 5)

for a in range(1, 5):
    ax.plot(x, func(x, a), 
        color=cmap(norm(a)))  # get color from color map 

# plot colorbar
fig.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
import math
from pathlib import Path
import json
```

```python
# some basic parameters
r = 150
w = 31.06
h = 25.79
offset = (1.32 - 2 * w, 2.59 + 4 * h)

# notch
s = 6
notch_coords = np.array([[2*r, r - s/2], [2*r, r + s/2], [2*r - math.sqrt(3)/2*s, r]])

map_fdir = Path().absolute().parent
df = pd.read_csv(map_fdir / 'dione_wafermap.csv')
wm = {}
for i, row in df.iterrows():
    wm[row['reticle']] = row['x']/1000, row['y']/1000 # unit mm
```

```python
# measured data json file
root = Path().resolve().parent / 'Dione_nit_wg_loss' / 'nit_loss_test_cells_BO_wafer2'
fname = 'loss_0.6um.json'

with open(root / fname, 'r') as f:
    loss_data = json.load(f)

loss_values = list(loss_data.values())
vmax = np.max(loss_values)
vmin = np.min(loss_values)
```

```python
# get a colormap
viridis = mpl.colormaps['viridis'].resampled(256)

# create all reticles
rets = {}
for id in range(len(wm)):
    # if ii in in_coord_trans:
    #     ret = plt.Rectangle((r + ii[0] * w, r + ii[1] * h), w, h, facecolor='#ff7f0e', edgecolor = 'k', fill=True)
    # else:
    x, y = r + wm[id][0] + offset[0], r + wm[id][1] + offset[1]
    ret = plt.Rectangle((x, y), w, h, color='k', fill=False)
    rets[id] = x, y, ret
```

```python
def to_str(idx: tuple) -> str:
    return '(' + str(idx[0]) + ', ' + str(idx[1]) + ')'
```

```python
# try colormap
def plot_examples(colormaps):
    """
    Helper function to plot data with associated colormap.
    """
    np.random.seed(19680801)
    data = np.random.randn(30, 30)
    n = len(colormaps)
    fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
                            constrained_layout=True, squeeze=False)
    for [ax, cmap] in zip(axs.flat, colormaps):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
        fig.colorbar(psm, ax=ax)
    plt.show()
```

```python
viridis = mpl.colormaps['viridis'].resampled(256)
plot_examples([viridis])
```

```python
top = mpl.colormaps['Oranges_r'].resampled(128)
bottom = mpl.colormaps['Blues'].resampled(128)

newcolors = np.vstack((top(np.linspace(0, 1, 128)),
                       bottom(np.linspace(0, 1, 128))))
newcmp = ListedColormap(newcolors, name='OrangeBlue')
plot_examples([newcmp])
```
